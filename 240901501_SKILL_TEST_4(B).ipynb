```python
#Perform Basic Data Analysis on Product Inventory Using Pandas
import pandas as pd
import matplotlib.pyplot as plt

# Given inventory data
data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Inventory Data:")
print(df)

# 1. Total number of products
print("\nTotal Products:", len(df))

# 2. Average price
print("Average Price:", df['Price'].mean())

# 3. Maximum price
print("Highest Price:", df['Price'].max())

# 4. Minimum price
print("Lowest Price:", df['Price'].min())

# 5. Total quantity of products
print("Total Quantity:", df['Quantity'].sum())

# Plotting price of products
df.plot(x='Product', y='Price', kind='bar', title='Product Price Comparison')

plt.xlabel("Product")
plt.ylabel("Price")
plt.show()
```
