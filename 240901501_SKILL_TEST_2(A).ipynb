```python
#Analyze Daily Step Count Using NumPy
import numpy as np
import matplotlib.pyplot as plt

# Step counts for 30 days (example data)
steps = np.array([
5000, 6200, 7000, 8000, 9000, 4500, 3000, 10000, 7500, 6800,
7200, 8500, 9000, 6700, 6400, 7800, 8200, 7600, 7100, 6900,
8300, 9100, 8700, 6200, 5800, 7600, 8200, 9000, 9400, 8800
])

# 1. Key Statistics
total_steps = np.sum(steps)
average_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print("Total Steps:", total_steps)
print("Average Steps per Day:", average_steps)
print("Maximum Steps:", max_steps)
print("Minimum Steps:", min_steps)

# 2. Identify High Activity Days
threshold = 8000
print("\nHigh Activity Days (Steps > 8000):")

for i in range(len(steps)):
    if steps[i] > threshold:
        print("Day", i+1, ":", steps[i], "steps")

# 3. Trend Analysis (increase or decrease)
print("\nDaily Trend Analysis:")

for i in range(1, len(steps)):
    if steps[i] > steps[i-1]:
        print("Day", i, "to Day", i+1, ": Increase")
    elif steps[i] < steps[i-1]:
        print("Day", i, "to Day", i+1, ": Decrease")
    else:
        print("Day", i, "to Day", i+1, ": No Change")

# 4. Visualization of Steps
days = np.arange(1, 31)

plt.plot(days, steps, marker='o')
plt.title("Daily Step Count Over a Month")
plt.xlabel("Day")
plt.ylabel("Steps")
plt.show()
```
