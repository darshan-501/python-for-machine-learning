```python
#Analyze Student Grades Using NumPy
import numpy as np
import matplotlib.pyplot as plt

# Marks of 30 students (Math, Science, English)
marks = np.random.randint(40,100,(30,3))

# Subject names
subjects = ["Math","Science","English"]

# Subject-wise statistics
average = np.mean(marks,axis=0)
highest = np.max(marks,axis=0)
lowest = np.min(marks,axis=0)

print("Subject Wise Statistics")

for i in range(3):
    print(subjects[i])
    print("Average:",average[i])
    print("Highest:",highest[i])
    print("Lowest:",lowest[i])
    print()

# Total marks of each student
total = np.sum(marks,axis=1)

# Identify top-performing student
top_student = np.argmax(total)

print("Top Performing Student Index:",top_student+1)
print("Total Marks:",total[top_student])

# Visualization
plt.hist(marks[:,0],alpha=0.5,label="Math")
plt.hist(marks[:,1],alpha=0.5,label="Science")
plt.hist(marks[:,2],alpha=0.5,label="English")

plt.title("Grade Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.legend()

plt.show()
```
