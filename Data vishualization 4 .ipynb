```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px

# seaborn style
sns.set()

# Correct data (same length)
Xpoints = np.array([0, 20, 120, 200, 250, 300, 350])
Ypoints = np.array([10, 80, 100, 120, 150, 180, 250])

# Matplotlib + Seaborn Line Plot
plt.figure(figsize=(8,5))
plt.plot(Xpoints, Ypoints, marker='o')
plt.title("Sample Line Plot")
plt.xlabel("X Points")
plt.ylabel("Y Points")
plt.grid(True)
plt.show()

# Plotly Interactive Plot
df = pd.DataFrame({"X": Xpoints, "Y": Ypoints})
fig = px.line(df, x="X", y="Y", markers=True, title="Interactive Plot")
fig.show()
```

```python
sns.scatterplot(data=df, x='X', y='Y')
plt.show()
```

```python
sns.barplot(data=df, x='X', y='Y')
plt.show()
```

```python
#3-D plot
X=np.array([1,2,6,8,12,14,20])
Y=np.array([3,8,1,10,12,16,18])
Z=np.array([5,1,9,4,11,7,15]) # Added Z data for 3D plot
Fig=plt.figure(figsize=(12,8))
ax=plt.axes(projection='3d') # Fixed typo and incorrect quote character
plt_3d=ax.scatter3D(X,Y,Z, c=Z, cmap='viridis') # Used X, Y, Z and added color for colorbar
plt.colorbar(plt_3d) # Corrected Plt to plt
plt.show()

#plotly package
df_3d = pd.DataFrame({'X': X, 'Y': Y, 'Z': Z}) # Created a DataFrame for plotly
fig_plotly = px.scatter_3d(df_3d, x='X', y='Y', z='Z', title="Interactive 3D Scatter Plot") # Used px.scatter_3d for a 3D plot
fig_plotly.show()
```
