```python
#Compute Total Impedance of an Electrical Circuit
import numpy as np

# Impedance values (complex numbers)
Z1 = 10 + 5j    # Resistor + Inductor
Z2 = 8 - 3j     # Resistor + Capacitor
Z3 = 5 + 2j     # Another component

# Series impedance
Z_series = Z1 + Z2 + Z3

# Parallel impedance
Z_parallel = 1 / (1/Z1 + 1/Z2 + 1/Z3)

print("Series Impedance:", Z_series)
print("Parallel Impedance:", Z_parallel)
```
