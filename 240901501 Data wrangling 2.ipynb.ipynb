```python
import pandas as pd
#Assign data
data={'Name':['Sai','Yuvasri','Poovannan',
'Sabharish','Sangamithra','Rithika','Rubasri'],
'Age':[20,18,19,20,18,17,19],
'Gender':['M','F','M','M','F','F','F'],
'Marks':[90,76,'NaN',74,65,'NaN',71]}
#convert into datafram
df=pd.DataFrame(data)
#display data
print("Original DataFrame:\n",df)

# Convert 'Marks' to numeric, coercing non-numeric values to NaN
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')

# Calculate the mean of the 'Marks' column, ignoring NaN values
avg_marks = df['Marks'].mean()

# Replace NaN values in 'Marks' with the calculated average
df['Marks'] = df['Marks'].fillna(avg_marks)

#Display data
print("\nReplacing NaN with average marks:\n",df)
```

```python
#Filter top scoring students
df=df[df['Marks']>=75]

#Remove age row
df=df.drop(['Age'],axis=1)

#Display data
print("\nFiltering and Dropping func:\n",df)
```
