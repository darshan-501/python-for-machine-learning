```python
import numpy as np

# Student names
names = ["Arun","Bala","Charan","Divya","Esha","Farhan","Gokul","Hari","Isha","John",
         "Kavin","Lakshmi","Manoj","Nila","Om","Priya","Rahul","Sneha","Tarun","Vijay"]

# Scores of 20 students
scores = np.array([45,78,56,89,90,67,34,23,88,76,54,61,72,48,59,92,81,39,64,70])

# 1. Overall Class Performance
print("Average Score:", np.mean(scores))
print("Highest Score:", np.max(scores))
print("Lowest Score:", np.min(scores))

print("\nStudent Analysis")

# 2. Pass/Fail and 3. Grade
for i in range(len(scores)):
    score = scores[i]

    # Pass / Fail
    if score >= 50:
        result = "Pass"
    else:
        result = "Fail"

    # Grade
    if score >= 90:
        grade = "A"
    elif score >= 75:
        grade = "B"
    elif score >= 60:
        grade = "C"
    elif score >= 50:
        grade = "D"
    else:
        grade = "F"

    print(names[i], "- Score:", score, "|", result, "| Grade:", grade)
```
